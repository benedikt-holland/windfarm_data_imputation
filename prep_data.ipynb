```python
import pandas as pd
```

```python
data = pd.read_csv('turbine_data.csv')
data['Hour'] = data['Tmstamp'].astype(str).str.split(":").str[0].astype(int)
data['Minute'] = data['Tmstamp'].astype(str).str.split(":").str[1].astype(int)
data['T'] = ((data['Day'] - 1) * 144) + (data['Hour'] * 6) + (data['Minute'] // 10)
data = data.sort_values(['TurbID', 'T'])
data.head()
```

```python
data['Patv'] = data.groupby('TurbID')['Patv'].transform(
    lambda series: series.interpolate(method='linear', limit_direction='both')
)
data.head()
```

```python
data.to_csv('data_interpolated.csv', index=False)
```
